```python
import matplotlib.pyplot as plt
import numpy as np  # for efficient array operations

# Define symbolic constants (replace with actual values if known)
C_b = 1000  # Base cost
beta = 50  # Other cost

# Sample data (replace with your actual data)
num_cargo_types = 2
demand = np.array([100, 50])  # Number of nodes for each cargo type
loading_cost = np.array([2, 3])  # Cost per unit for loading each cargo type
loading_time = np.array([1, 2])  # Time per unit for loading each cargo type
total_cycle_time = 20  # Total order cycle time
unloading_time = np.array([0.5, 1])  # Time per unit for unloading each cargo type
unloading_cost = np.array([1, 1.5])  # Cost per unit for unloading each cargo type

# Generate time points (assuming time might affect base cost)
time_points = np.arange(1, total_cycle_time + 1)  # Array of time points

def total_cost(t):
  """
  This function calculates the total cost (CT) based on the given equations:

  CT = C_b + sum(N_n * C_Ln * T_Ln) + (T_oct - sum(T_Ln + T_Un)) * sum(C_n) + sum(N_n * C_Un * T_Un) + beta

  where:
      CT - Total cost
      C_b - Base cost
      N_n - Number of nodes for cargo type n
      C_Ln - Loading cost per unit for cargo type n
      T_Ln - Loading time per unit for cargo type n
      T_oct - Total order cycle time
      C_n - Cost per unit of remaining time for cargo type n (assumed constant here)
      T_Un - Unloading time per unit for cargo type n
      C_Un - Unloading cost per unit for cargo type n
      beta - Other cost
  """

  adjusted_base_cost = C_b + t  # Assuming base cost increases with time
  loading_cost_term = np.sum(demand * loading_cost * loading_time)
  unloading_cost_term = np.sum(demand * unloading_cost * unloading_time)
  effective_cycle_time = total_cycle_time - np.sum(loading_time + unloading_time)
  total_cost_n = adjusted_base_cost + loading_cost_term + effective_cycle_time * np.sum(demand * unloading_cost) + unloading_cost_term + beta
  return total_cost_n

# Calculate total cost for each time point
costs = np.array([total_cost(t) for t in time_points])

# Plot the cost vs time graph
plt.plot(time_points, costs)
plt.xlabel("Time")
plt.ylabel("Total Cost")
plt.title("Cost vs. Time")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x-axis (adjust as needed)
x_range = np.linspace(1, 10, 100)  # Creates 100 points between 1 and 10 (inclusive)

# Since x is inversely proportional to y, calculate y as the reciprocal of x
y = 1 / x_range

# Plot the graph
plt.plot(x_range, y)

# Set axis labels and title
plt.xlabel("X (Independent Variable)")
plt.ylabel("Y (Dependent Variable)")
plt.title("Inverse Proportional Relationship")

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
